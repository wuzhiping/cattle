```python
import asyncio
import uuid
import json
import time
from temporalio.client import Client

from utils.blowfish_ctr import encrypt
from utils.get_config import Config


async def main():
    payload = {
        "dir": "./demo",
        "question": "当前工作目录中的文件"
    }
    
    # 转换并拼接at+时间戳
    result = json.dumps(payload, ensure_ascii=False) + f"at{int(time.time())}"
    print("result：", result)
    print(type(result))

    # 获取配置
    config = await Config().fetch()
    _key = config.get("temporal").get("task_queue")
    _addr = config.get("temporal").get("addr")
    _ns = config.get("temporal").get("namespace")

    # 加密
    enc = encrypt(str(_key), result)
    print("加密：", enc)
    print(type(enc))

    # 运行
    client = await Client.connect(_addr, namespace=_ns)
    result = await client.execute_workflow(
        "OPflow",
        enc,
        id=f"say-hello-workflow-{uuid.uuid4()}",
        task_queue=str(_key),
    )
    print("Workflow result:", result)

if __name__ == "__main__":
    await main()
```

```python
from Crypto.Cipher import Blowfish
from Crypto.Util.Padding import pad, unpad
import hashlib

pwd, salt = '123456789', b'salt'
key = hashlib.scrypt(pwd.encode(), salt=salt, n=16384, r=8, p=1, dklen=24)[:16]
iv = bytes(8)

def encrypt(data):
    return Blowfish.new(key, Blowfish.MODE_CBC, iv).encrypt(pad(data.encode(), 8)).hex()

def decrypt(encrypted_hex):
    return unpad(Blowfish.new(key, Blowfish.MODE_CBC, iv).decrypt(bytes.fromhex(encrypted_hex)), 8).decode()

# 测试
enc = encrypt("jupyter, hello world!")
print(f"加密: {enc}")
print(f"解密: {decrypt(enc)}")
```

```python
from Crypto.Cipher import AES  # 改1：Blowfish → AES
from Crypto.Util.Padding import pad, unpad
import hashlib

pwd, salt = '123456789', b'salt'
key = hashlib.scrypt(pwd.encode(), salt=salt, n=16384, r=8, p=1, dklen=24)  # 改2：24字节正好给AES-192
iv = bytes(16)  # 改3：AES IV 16字节（Blowfish是8字节）

def encrypt(data):
    return AES.new(key, AES.MODE_CBC, iv).encrypt(pad(data.encode(), 16)).hex()  # 改4：块大小16字节

def decrypt(encrypted_hex):
    return unpad(AES.new(key, AES.MODE_CBC, iv).decrypt(bytes.fromhex(encrypted_hex)), 16).decode()  # 改5：块大小16字节

# 测试
# enc = decrypt("f0b1928e05f4f53b4b17cd89780978bc")
# print(f"解密: {enc}")

# 测试
enc = encrypt("zhengqijun")
print(f"加密: {enc}")
print(f"解密: {decrypt(enc)}")
```
