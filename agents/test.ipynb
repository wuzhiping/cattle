```python
import asyncio
import uuid
from temporalio.client import Client

async def main():
    payload = {
        "dir": "./demo",
        "question": "当前工作目录中的文件"
    }
    client = await Client.connect("localhost:7233")
    result = await client.execute_workflow(
        "OPflow",
        payload,
        id=f"say-hello-workflow-{uuid.uuid4()}",
        task_queue="_device_",
    )
    print("Workflow result:", result)

if __name__ == "__main__":
    await main()
```
